```python
plt.figure(1)
plt.subplot(2,2,1)
df['engine-type'].value_counts().plot(figsize=(10,8),kind='bar',color='purple')
plt.title('Engine Type Frequence Diagram')
plt.ylabel('No of Engine Type')
plt.xlabel('Engine Type')
plt.grid(True)

plt.subplot(2,2,2)
df['drive-wheels'].value_counts().plot(figsize=(10,8),kind='bar',color='green')
plt.title("Drive Wheels Frequence Diagram")
plt.ylabel("No of Drive Wheels Type")
plt.xlabel("Drive Wheels Type")
plt.grid(True)

plt.subplot(2,2,3)
df['fuel-type'].value_counts().plot(figsize=(10,8),kind='bar',color='orange')
plt.title('Fuel Type Frequence Diagram')
plt.ylabel('No of Fuel Type')
plt.xlabel('Fuel Type')
plt.grid(True)

plt.subplot(2,2,4)
df['body-style'].value_counts().plot(figsize=(10,8),kind='bar',color='Pink')
plt.title('Body Type Frequence Diagram')
plt.ylabel('No of Body Style Type')
plt.xlabel('Body Style Type')
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Sub plot

```python
plt.figure(1)
plt.subplot(2,3,1)
df['engine-type'].value_counts().plot(kind='bar',color='r',figsize=(20,14))
plt.title('Engine')
plt.xlabel('Engine-Type')
plt.ylabel('No. of Engines')

plt.subplot(2,3,2)
df['fuel-type'].value_counts().plot(kind='bar',color='g',figsize=(20,14))
plt.title('Fuel-Type')
plt.xlabel('Fuel')
plt.ylabel('no. of cars')

plt.subplot(2,3,3)
df['drive-wheels'].value_counts().plot(kind='bar',color='b',figsize=(20,14))
plt.title('Drive-Wheels')
plt.xlabel('Wheel name')
plt.ylabel('Quantity of wheels')

plt.subplot(2,3,4)
df['body-style'].value_counts().plot(kind='bar',color='pink',figsize=(20,14))
plt.title('Body-Style')
plt.xlabel('Styles')
plt.ylabel('No. of cars')

plt.subplot(2,3,5)
df['engine-location'].value_counts().plot(kind='bar',color='black',figsize=(20,14))
plt.title('Engine location')

plt.subplot(2,3,6)
df['make'].value_counts().plot(kind='pie',figsize=(20,14))
plt.title('Make')

plt.show()
```
